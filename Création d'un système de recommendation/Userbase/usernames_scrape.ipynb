On va maintenant essayer de récuperer des user_id en obtenant les listes d'utilisateurs à partir des mangas les plus connus.

```python
mangas = mangas[['manga_id','title','scored_by']]
mangas = mangas.sort_values('scored_by', ascending=False).head(50)
mangas
```

```python
cols = ['username']

def scrape_usernames():
    """
    Pour chaque manga dans la listes des 50 mangas les plus populaires définie au dessus, prend les 50 premières pages
    d'utilisateurs ayant récemment modifié leur statut sur ce manga.
    
    """
    pages = 50
    count = 0
    all_users = pd.Series(dtype='object')
    start_time = time.time()
    for manga_id in ([mangas['manga_id'][5]]):
        for page in range(1,pages):
            apiUrl = f'https://api.jikan.moe/v4/manga/{manga_id}/userupdates?page={page}'
            users = pd.Series(dtype='object')
            # API call with retries
            tries = 0
            while tries < 3:
                response = requests.get(apiUrl)

                # Vérifie que la 
                if response.status_code == 200:
                    break  # Successful response, exit the retry loop

                tries += 1

            if response.status_code == 200:
                # Parse the JSON response
                data = response.json()

                # Check if the 'data' field exists in the response
                if ('data' in data):
                    # Extract the user details
                    if len(data['data'])>0:
                        for i in range(len(data['data'])):
                            user = pd.Series(data['data'][i]['user']['username'])                                  
                            users = pd.concat([users, user])
                            count += 1

                        # Create a dictionary for the user
                        # Append the user dictionary to the list
                        all_users = pd.concat([all_users,users])

                    else:
                        print(f'No user data found for ID: {manga_id}')

                # else:
                #     print(f'Error occurred while fetching user data for ID: {user_id}')

    end_time = time.time()
    print(f'The extraction of {count} users took {end_time-start_time} seconds')
    return all_users
```
