```python
import requests
import pandas as pd
import numpy as np
import time

token = pd.read_json('../Userbase/token.json', orient='index').T['access_token'][0]
token

headers ={"Authorization": f'Bearer {token}'}
```

```python
users = pd.read_csv('../Userbase/first_scrap.csv')
users
```

```python
fields = 'node{id, title, mean}, list_status{score, status}'
def scrape_list(user):
    npage = []
    start_time = time.time()
    url = f"https://api.myanimelist.net/v2/users/{user}/mangalist?fields={fields}&sort=list_score"
    
    # API call with retries
    tries = 0
    while tries < 3:
        time.sleep(0.3)
        response = requests.get(url, headers=headers)

        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            break  # Successful response, exit the retry loop

        tries += 1
                
    if response.status_code == 200:
        data = response.json()
        if 'data' in data:
            npage = iter_page(user,data)
    print(f'Temps de traitement: {time.time() - start_time} secondes, User {i}')
    return npage
```

```python
def iter_page(user, data):
    agg_list = clean_dict(data)
    page = 0
    while ('next' in data['paging']) & (page < 80):
        page += 1
        urlx = data['paging']['next']
        tries = 0
        while tries < 3:
            time.sleep(0.3)
            response = requests.get(urlx, headers=headers)

            # Check if the request was successful (status code 200)
            if response.status_code == 200:
                break  # Successful response, exit the retry loop

            tries += 1
            
        if response.status_code == 200:    
            data = response.json()
            nlist = clean_dict(data)
            agg_list.extend(nlist)
    return agg_list

def clean_dict(data):
    nlist = []
    for i in range(len(data['data'])):
        manga = data['data'][i]['node']
        if 'main_picture' in manga:
            del manga['main_picture']
        status = data['data'][i]['list_status']
        delete = ['is_rereading','num_volumes_read','num_chapters_read', 'updated_at', 'start_date','finish_date']
        for field in delete:
            if field in status:
                del status[field]
        manga.update(status)
        nlist.append(manga)
    return nlist
```

```python
user_lists = []
st_time = time.time()
for i in range(0,3500):
    ls = scrape_list(users[i])
    user_lists.append([users[i], ls])
    print(f'Temps écoulé : {time.time()-st_time} secondes')
print(f'Temps total : {time.time()-st_time} secondes')
```

```python
user_df = pd.DataFrame.from_dict(user_lists)
user_df.columns = ['Users','infos']
```

On veut maintenant séparer les informations dans la colonne infos pour créer toutes les lignes pour tous les utilisateurs.

```python
def dict_to_df(index):
    pos = user_df['infos'][index]
    user = pd.Series(user_df['Users'][index], range(len(user_df['infos'][index])))
    if user_df['infos'][index] != []:
        mangas = pd.DataFrame.from_dict(user_df['infos'][index])
        df = pd.concat([user,mangas], axis=1)
        return df
```

```python
final_df = []
for i in range(10):
    fdfx = pd.DataFrame()
    for j in range((0)*3500,(1)*3500):
        if j%500 ==0:
            print(f'User n°{j}')
        temp_df = dict_to_df(j)
        fdfx = pd.concat([fdfx, temp_df])
    final_df.append(fdfx)
```

```python
full_df = pd.DataFrame(columns=final_df[0].columns)
for i in range(len(final_df)):
    full_df = pd.concat([full_df,final_df[i]])
```
