La première approche que je vais essayer est de mettre en place un modèle de Collaborative Filtering, ce qui consiste à créer une matrice contenant pour chaque utilisateur l'ensemble des notes qu'il a attribué (0 si pas de notes) sur l'ensemble des mangas que l'on considère. Ce modèle se basera uniquement sur les notes attribuées, et on étudiera par la suite un modèle faisant appel au Content Based Filtering pour comparer les deux méthodes.

```python
# On commence par importer les librairies et les fichiers de données.
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras

users = pd.read_csv('../User lists/all_user_lists_v2.csv')
mangas = pd.read_csv('sorted_manga_list.csv')
```

```python
mangas
```

```python
matrix = users[['user','title','score']].pivot_table(index='user', columns='title', values='score')
matrix.replace({np.nan : 0}, inplace=True)
matrix = matrix.astype('int32')
matrix
```

```python
Y = np.array(matrix).T
Y
```

```python
R = np.array(matrix.where(matrix == 0, other=1)).T
R
```

```python
num_mangas = 1067
num_users = 33818
```

```python
def cofi_cost_func_v(X, W, b, Y, R, lambda_):
    """
    Returns the cost for the content-based filtering
    Vectorized for speed. Uses tensorflow operations to be compatible with custom training loop.
    Args:
      X (ndarray (num_movies,num_features)): matrix of item features
      W (ndarray (num_users,num_features)) : matrix of user parameters
      b (ndarray (1, num_users)            : vector of user parameters
      Y (ndarray (num_movies,num_users)    : matrix of user ratings of movies
      R (ndarray (num_movies,num_users)    : matrix, where R(i, j) = 1 if the i-th movies was rated by the j-th user
      lambda_ (float): regularization parameter
    Returns:
      J (float) : Cost
    """
    j = (tf.linalg.matmul(X, tf.transpose(W)) + b - Y)*R
    J = 0.5 * tf.reduce_sum(j**2) + (lambda_/2) * (tf.reduce_sum(X**2) + tf.reduce_sum(W**2))
    return J
```

```python
mylist = pd.read_csv('../Création d\'un système de recommendation/User lists/mylist.csv')
```

```python
mylist
```

```python
manga_list = pd.Series(matrix.columns)
my_scores = np.zeros(num_mangas)

for i in range(len(manga_list)):
    if mylist['title'].eq(manga_list[i]).any():
        my_scores[i] = mylist[mylist['title']==manga_list[i]]['score'].item()
    else:
        my_scores[i] = 0   
```

```python
def normalizeRatings(Y, R):
    """
    Preprocess data by subtracting mean rating for every movie (every row).
    Only include real ratings R(i,j)=1.
    [Ynorm, Ymean] = normalizeRatings(Y, R) normalized Y so that each movie
    has a rating of 0 on average. Unrated moves then have a mean rating (0)
    Returns the mean rating in Ymean.
    """
    Ymean = (np.sum(Y*R,axis=1)/(np.sum(R, axis=1)+1e-12)).reshape(-1,1)
    Ynorm = Y - np.multiply(Ymean, R) 
    return(Ynorm, Ymean)
```

```python
Y = np.c_[my_scores, Y]
R = np.c_[(my_scores != 0).astype(int), R]

Ynorm, Ymean = normalizeRatings(Y, R)
```

```python
#  Useful Values
num_movies, num_users = Y.shape
num_features = 100

# Set Initial Parameters (W, X), use tf.Variable to track these variables
tf.random.set_seed(1234) # for consistent results
W = tf.Variable(tf.random.normal((num_users,  num_features),dtype=tf.float64),  name='W')
X = tf.Variable(tf.random.normal((num_movies, num_features),dtype=tf.float64),  name='X')
b = tf.Variable(tf.random.normal((1,          num_users),   dtype=tf.float64),  name='b')

# Instantiate an optimizer.
optimizer = keras.optimizers.Adam(learning_rate=1e-1)
```

```python
iterations = 200
lambda_ = 1
for iter in range(iterations):
    # Use TensorFlow’s GradientTape
    # to record the operations used to compute the cost 
    with tf.GradientTape() as tape:

        # Compute the cost (forward pass included in cost)
        cost_value = cofi_cost_func_v(X, W, b, Ynorm, R, lambda_)

    # Use the gradient tape to automatically retrieve
    # the gradients of the trainable variables with respect to the loss
    grads = tape.gradient( cost_value, [X,W,b] )

    # Run one step of gradient descent by updating
    # the value of the variables to minimize the loss.
    optimizer.apply_gradients( zip(grads, [X,W,b]) )

    # Log periodically.
    if iter % 20 == 0:
        print(f"Training loss at iteration {iter}: {cost_value:0.1f}")
```

```python
my_rated = [i for i in range(len(my_scores)) if my_scores[i] > 0]

# Make a prediction using trained weights and biases
p = np.matmul(X.numpy(), np.transpose(W.numpy())) + b.numpy()

#restore the mean
pm = p + Ymean

my_predictions = pm[:,0]

# sort predictions
ix = tf.argsort(my_predictions, direction='DESCENDING')
```

```python
manga_array = np.array(manga_list)
manga_array
```

```python
for i in range(30):
    j = ix[i]
    if j not in my_rated:
        print(f'Predicting rating {my_predictions[j]:0.2f} for movie {manga_array[j]}')
```

```python
print('\n\nOriginal vs Predicted ratings:\n')
for i in range(len(my_scores)):
    if my_scores[i] > 0:
        print(f'Original {my_scores[i]}, Predicted {my_predictions[i]:0.2f} for {manga_array[i]}')
```
